```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4])
arr2 = np.arange(0,10,2)
arr3 = np.random.randint(0,100,5)

print("Array 1:", arr1)
print("Array 2:", arr2)
print("Random Array:", arr3)

print("Addition:", arr1 + 5)
print("Multiplication:", arr1 * 2)
print("Mean:", np.mean(arr3))
print("Max:", np.max(arr3))
print("Reshape:", arr2.reshape(1, 5))
```

```python
import pandas as pd
```

```python
data = {
    "Name" : ["Harsh", "Ashwini", "Amit"],
    "Age" : [22,17,35],
    "City" : ["Ranchi", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)
print("First two rows:\n", df.head(2))
print("Describe:\n", df.describe())
```

```python
print("Age > 20 : ", df[df["Age"] > 20])
```
